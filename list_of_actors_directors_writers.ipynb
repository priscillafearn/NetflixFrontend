# LIST OF ACTORS (FAMOUS / NON-FAMOUS)

```python
import pandas as pd
import numpy as np
url = "/Users/davidguenoun/code/davidguenoun/wagon_final_project_netflix/Netflix/raw_data/merged_movies_by_index.csv"
df = pd.read_csv(url)
```

```python
df = df[["Actors"]].dropna()
```

```python
df["Actors"] = df["Actors"].apply(lambda x : x.split(","))
```

```python
df.head()
```

```python
actor_list = []
for index, row in df.iterrows():
    actor_list += row["Actors"]
```

```python
list_famous_actor = []
for actor in actor_list:
    if actor_list.count(actor) > 20:
        list_famous_actor.append(actor)
        
```

```python
list_famous_actor = pd.Series(list_famous_actor).drop_duplicates().tolist()[0:5]
```

```python
list_famous_actor
```

```python
list_non_famous_actor = []
for actor in actor_list:
        if actor_list.count(actor) < 3:
            list_non_famous_actor.append(actor)

pd.Series(list_non_famous_actor).drop_duplicates().tolist()
```

```python
list_non_famous_actor = pd.Series(list_non_famous_actor).drop_duplicates().tolist()[0:5]
```

```python
list_non_famous_actor
```
