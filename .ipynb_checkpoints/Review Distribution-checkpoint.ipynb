```python
import pandas as pd
import numpy as np
url = "/Users/davidguenoun/code/davidguenoun/wagon_final_project_netflix/Netflix/raw_data/merged_movies_by_index.csv"
df = pd.read_csv(url)
```

```python
df["Writer"][9]
```

```python
df = df[["Director"]].dropna()
```

```python
df[["Director"]]
```

```python
director_list = []
for director in df["Director"]:
    director_list.append(director)
director_list
```

```python
famous_director_list = []
for director in director_list:
    if director_list.count(director) > 20:
        famous_director_list.append(director)

famous_director_list = pd.Series(famous_director_list).drop_duplicates().tolist()[0:5]
famous_director_list
```

```python
non_famous_director_list = []
for director in director_list:
    if director_list.count(director) < 3:
        non_famous_director_list.append(director)

non_famous_director_list = pd.Series(non_famous_director_list).drop_duplicates().tolist()[0:5]
non_famous_director_list
```

```python
director_list_final = famous_director_list + non_famous_director_list
```

# Display avg review score distribution

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
url = "/Users/davidguenoun/code/davidguenoun/wagon_final_project_netflix/Netflix/raw_data/merged_movies_by_index.csv"
df = pd.read_csv(url)
x = df["avg_review_score"]
```

```python
sns.distplot(df["avg_review_score"],hist = False, color = "r")
plt.axvline(x=df["avg_review_score"][28], c="b");
```

```python
percentile = round(1 - (df["avg_review_score"] > df["avg_review_score"][28]).mean(),2)
```

```python
percentile = "{:.0%}".format(percentile)
```

```python
sentence = f'This movie has a better review_score than {percentile} of movies on Netflix'
```

```python
sentence
```

```python



def review_score(x):
    
    
    percentile = round(1 - (df["avg_review_score"] > df["avg_review_score"][x]).mean(),2)
    percentile = "{:.0%}".format(percentile)
    sentence = f'This movie has a better review_score than {percentile} of movies on Netflix'
    vertical_line = plt.axvline(x=df["avg_review_score"][x], c="b");
    distribution = sns.distplot(df["avg_review_score"],hist = False, color = "r");
    
    return vertical_line, distribution,sentence

    
    
    
    
    
```

```python
review_score(40)
```
